```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
#define image size and batch size
IMG_SIZE=224
BATCH_SIZE=32
```

```python
#define data generators for train, validatiion,and test sets
train_datagen=ImageDataGenerator(rescale=1./255,validation_split=0.2)
train_generator=train_datagen.flow_from_directory("/content/drive/MyDrive/DATASET/Brain_Tumor_Detection/train",
target_size=(IMG_SIZE,IMG_SIZE),
batch_size=BATCH_SIZE,
class_mode='binary',
subset='validation'
)
validate_generator=train_datagen.flow_from_directory("/content/drive/MyDrive/DATASET/Brain_Tumor_Detection/train",
target_size=(IMG_SIZE,IMG_SIZE),
batch_size=BATCH_SIZE,
class_mode='binary',
subset='validation'
)
test_datagen=ImageDataGenerator(rescale=1./255)
test_generator=train_datagen.flow_from_directory("/content/drive/MyDrive/DATASET/Brain_Tumor_Detection/train",
target_size=(IMG_SIZE,IMG_SIZE),
batch_size=BATCH_SIZE,
class_mode='binary',
subset='validation'
)
```

```python
#define the model
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
model=keras.Sequential([

      layers.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3)),
      layers.MaxPooling2D((2,3)),
      layers.Conv2D(64,(3,3),activation='relu'),
      layers.MaxPooling2D((2,2)),
      layers.Conv2D(128,(3,3),activation='relu'),
      layers.MaxPooling2D((2,2)),
      layers.Flatten(),
      layers.Dense(128,activation='relu'),
      layers.Dense(1,activation='sigmoid')
])
```

```python
#complie the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(train_generator,validation_data=validate_generator,epochs=5)
```

```python
model.save("Model.h5","label.txt")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
#load the saved model
model=load_model('/content/Model.h5')
#load and preproocess the test image
test_image_path='/content/drive/MyDrive/DATASET/Brain_Tumor_Detection/test/pred15.jpg'
img = image.load_img(test_image_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /=225
prediction = model.predict(img_array)
if prediction < 0.5:
  print("prediction: No tumor (probability:", prediction[0][0], ")")
else:
    print("prediction: Tumor present (probability:",prediction[0][0], ")")
```
