```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
car_crashes=sns.load_dataset("car_crashes")
car_crashes.to_csv('s.csv')
d=pd.read_csv('/content/s.csv')
print(car_crashes)
sns.jointplot(x="total",y="speeding",data=car_crashes)
plt.title("caar crashes ")
plt.xlabel("total")
plt.ylabel("due to speed")
plt.show()
```
